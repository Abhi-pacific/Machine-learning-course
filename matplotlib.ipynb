```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
heart_disease = pd.read_csv(r"D:\ML U\Dataset\heart-disease.csv",encoding='utf-8')
```

```python
heart_disease.head()
```

# creating the graph with the OO method.

```python
# creating the subplot
fig, (ax1,ax2) = plt.subplots(nrows = 2,
                              ncols= 1,
                              figsize=(10,10))

# adding the data in the graph
scatter = ax2.scatter(
    x = heart_disease['thalach'],
    y = heart_disease['age'],
    c = heart_disease['target']
)

# adding the labels in the graph
ax2.set(xlabel = 'Thalach',
       ylabel = 'Age',
       title = 'Graph for the Thalach as per the age')

# creating the labels.
ax2.legend(*scatter.legend_elements(),title = 'Target')

# CREATING THE MEAN LINE.
ax2.axhline(heart_disease['age'].mean())
plt.show()

# Setting the main title in this graph.
```
